# DataSet Terms Explanation

1. **Age**: Age is a well-known risk factor for heart disease. As individuals age, the risk of developing heart-related issues increases. This is partly because aging can lead to changes in the heart and blood vessels, including atherosclerosis (build-up of plaque in the arteries), which can increase the risk of heart attacks and other cardiovascular diseases.

2. **Gender**: Gender differences exist in the prevalence and presentation of heart disease. Men tend to have a higher risk of heart disease at younger ages compared to women. However, after menopause, women's risk catches up to that of men. Estrogen is believed to offer some protective effects against heart disease in premenopausal women. Women are more likely to experience atypical symptoms of heart disease, which can lead to underdiagnosis and delayed treatment.

3. **Chest Pain Type**: The type of chest pain (angina) experienced can provide clues about the underlying cause. For example, stable angina typically occurs during physical exertion or emotional stress and is relieved by rest or medication. Unstable angina, on the other hand, may occur at rest and can signal a more serious condition like a heart attack.

4. **Resting Blood Pressure**: High blood pressure (hypertension) is a major risk factor for heart disease. It can damage the arteries over time, making them more susceptible to atherosclerosis. Hypertension increases the workload on the heart and can lead to heart failure, heart attack, stroke, and other complications.

5. **Serum Cholesterol**: High levels of low-density lipoprotein (LDL) cholesterol ("bad" cholesterol) and low levels of high-density lipoprotein (HDL) cholesterol ("good" cholesterol) are associated with an increased risk of atherosclerosis and heart disease. LDL cholesterol can deposit in the walls of arteries, forming plaques that narrow the arteries and restrict blood flow to the heart.

6. **Fasting Blood Sugar**: High fasting blood sugar levels can indicate diabetes or prediabetes, both of which are significant risk factors for heart disease. Diabetes accelerates the development of atherosclerosis and increases the risk of heart attack, stroke, and other cardiovascular complications.

7. **Resting Electrocardiographic Results (Resting ECG)**: An abnormal ECG can suggest various heart conditions, including coronary artery disease, arrhythmias, and heart muscle abnormalities. Certain ECG patterns, such as ST-segment changes, can indicate acute myocardial infarction (heart attack).

8. **Maximum Heart Rate Achieved**: The maximum heart rate achieved during exercise testing can provide information about cardiovascular fitness. A lower maximum heart rate may indicate reduced cardiac function and increased cardiovascular risk.

9. **Exercise Induced Angina**: Angina induced by exercise (exertional angina) suggests underlying coronary artery disease. It occurs when the heart muscle doesn't receive enough blood and oxygen during physical activity. Exertional angina is a warning sign of potential heart attack risk.

10. **ST Depression Induced by Exercise**: ST-segment depression observed during exercise testing indicates myocardial ischemia (insufficient blood flow to the heart muscle). It is a strong predictor of coronary artery disease and can signal an increased risk of heart attack.

11. **Slope of Peak Exercise ST Segment**: The slope of the ST segment during exercise testing provides additional information about the severity and extent of myocardial ischemia. A downsloping ST segment is more indicative of significant coronary artery disease and a higher risk of adverse cardiac events.

12. **Number of Major Vessels Colored by Flourosopy**: The number of major coronary vessels with significant stenosis (narrowing) detected by fluoroscopy is a direct indicator of the severity of coronary artery disease. Multiple vessel involvement increases the risk of heart attack and other cardiovascular events.

13. **Thalassemia Type**: Thalassemia is a genetic blood disorder that affects hemoglobin production. Certain types of thalassemia, particularly thalassemia major, can lead to heart complications such as cardiomyopathy (disease of the heart muscle) and heart failure.

14. **Heart Disease (Target)**: This is the presence or absence of heart disease, typically defined as the presence of significant coronary artery disease, myocardial infarction (heart attack), or other related conditions. The other features mentioned above contribute to the risk of developing heart disease and can help in the diagnosis and management of patients with suspected heart conditions.


## Causes of Heart Attack


1. **Aging**: As individuals age, the risk of developing heart-related issues increases due to changes in the heart and blood vessels.

2. **Hypertension (High Blood Pressure)**: High blood pressure damages the arteries over time, increasing the risk of atherosclerosis and heart attack.

3. **High Cholesterol Levels**: Elevated levels of LDL cholesterol ("bad" cholesterol) contribute to the formation of arterial plaques, narrowing the arteries and increasing the risk of heart attack.

4. **Diabetes**: Diabetes accelerates the development of atherosclerosis and increases the risk of heart attack and other cardiovascular complications.

5. **Obesity**: Excess body weight, particularly abdominal obesity, is associated with hypertension, diabetes, and dyslipidemia, all of which are major risk factors for heart attack.

6. **Sedentary Lifestyle**: Lack of physical activity contributes to obesity, hypertension, and dyslipidemia, increasing the risk of heart attack.

7. **Smoking**: Tobacco smoke contains chemicals that damage blood vessels and promote atherosclerosis, significantly increasing the risk of heart attack.

8. **Genetic Factors**: Family history of heart disease and genetic predispositions can increase the likelihood of developing heart attack risk factors such as hypertension, high cholesterol, and diabetes.

9. **Unhealthy Diet**: Poor dietary habits high in saturated fats, trans fats, cholesterol, and sodium contribute to hypertension, dyslipidemia, obesity, and diabetes, all of which increase heart attack risk.

10. **Stress**: Chronic stress can lead to unhealthy coping mechanisms such as overeating, smoking, and physical inactivity, exacerbating heart attack risk factors.

11. **Alcohol Consumption**: Excessive alcohol consumption can raise blood pressure, contribute to obesity, and increase triglyceride levels, all of which are associated with higher heart attack risk.

12. **Sleep Apnea**: Sleep apnea is associated with hypertension, obesity, and insulin resistance, all of which increase the risk of heart attack.

13. **Poor Dental Health**: Gum disease (periodontitis) has been linked to an increased risk of heart disease and heart attack due to the inflammation and bacterial infections it causes.

14. **Drug Abuse**: Illicit drug use, particularly stimulants like cocaine and methamphetamine, can cause arterial spasms, increased heart rate, and elevated blood pressure, raising the risk of heart attack.

15. **Undiagnosed or Untreated Heart Conditions**: Conditions such as coronary artery disease, arrhythmias, heart valve disorders, and congenital heart defects can significantly increase the risk of heart attack if left untreated or undiagnosed.

```python
def has_categorical_column(df):
    """Check if the DataFrame has at least one categorical column."""
    for col in df.columns:
        if df[col].dtype == 'object':
            return True
    return False

# Example usage
if has_categorical_column(df):
    print("The DataFrame contains at least one categorical column.")
else:
    print("The DataFrame does not contain any categorical column.")
```

```python
def plot_df(df):
    
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    categorical_cols = df.select_dtypes(include=['object'])
    
    
    for col in numerical_cols:
        plt.figure(figsize=(10,6))
        sns.histplot(df[col], kde=True)
        plt.title(f"Distribution plot of {col}")
        plt.xlabel(f"{col}")
        plt.xticks(rotation=45)
        plt.show()
        
        
        
    for col in categorical_cols:
        plt.figure(figsize=(10, 6))
        sns.countplot(x=col, data=df)
        plt.title(f"Count plot of {col}")
        plt.xlabel(f"{col}")
        plt.xticks(rotation=45)
        plt.show()
        
        
plot_df(df)
```

```python
from sklearn.preprocessing import RobustScaler

X = df.drop('target', axis=1)
y = df['target']

scaler = RobustScaler()

X_scaled = scaler.fit_transform(X)

X_scaled.shape


X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
X_train.shape
```

```python
classifiers = {
    "Gaussian NB": GaussianNB(),
}

# Initialize variables to track best model and its accuracy
best_model = None
best_accuracy = 0

# Train and evaluate each classifier
for name, clf in classifiers.items():
    try:
        # Train the classifier
        clf.fit(X_train, y_train)
        
        # Predict on the test set
        y_pred = clf.predict(X_test)
        
        # Calculate the accuracy
        accuracy = accuracy_score(y_test, y_pred)
        
        # Update best model if accuracy is higher
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model = (name, clf)
        
        print(f"{name}: Accuracy = {accuracy:.2f}")
        print("="*75)
    except Exception as e:
        print(f"{name}: Error during training or prediction - {e}")

# Print the best model and its accuracy
print(f"\nBest Model: {best_model[0]}, Accuracy = {best_accuracy:.2f}")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import RobustScaler

# Initialize variables to store the best fold's metrics
best_accuracy = 0
best_fold_index = None
best_classifier = None
best_predictions = None

# Initialize the Gaussian NB classifier with specified parameters
classifier = GaussianNB(priors=None, var_smoothing=1e-09)

# Define the number of folds (5 folds in this case)
n_splits = 11

# Initialize StratifiedKFold
stratified_kfold = StratifiedKFold(n_splits=n_splits)

# Loop through each fold
for fold_index, (train_index, val_index) in enumerate(stratified_kfold.split(X_scaled, y)):
    # Split the dataset into training and validation sets
    X_train_fold, X_val_fold = X_scaled[train_index], X_scaled[val_index]
    y_train_fold, y_val_fold = y[train_index], y[val_index]

    # Train the classifier on the training set
    classifier.fit(X_train_fold, y_train_fold)

    # Predict on the validation set
    y_val_pred = classifier.predict(X_val_fold)

    # Calculate the accuracy on the validation set
    accuracy = accuracy_score(y_val_fold, y_val_pred)
    print("Accuracy on fold", fold_index + 1, ":", round(accuracy, 3))

    # Check if current accuracy is higher than the best accuracy
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_fold_index = fold_index
        best_classifier = classifier
        best_predictions = y_val_pred

# Print the best fold index and accuracy
print("Best fold index:", best_fold_index + 1)
print("Best accuracy:", round(best_accuracy, 3))

# You can use best_classifier, best_predictions, and best_accuracy variables for further analysis or predictions.
```
