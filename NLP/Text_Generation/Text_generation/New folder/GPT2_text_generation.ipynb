```python
from transformers import GPT2Tokenizer, TFGPT2LMHeadModel
import tensorflow as tf
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Set the pad token ID to the EOS token ID
tokenizer.pad_token = tokenizer.eos_token

model = TFGPT2LMHeadModel.from_pretrained('gpt2')
```

```python
input_text = "Natural Language Processing is a subset of Machine"

input_ids = tokenizer.encode(input_text, return_tensors='tf')
```

```python
attention_mask = tf.ones_like(input_ids)



output = model.generate(
    input_ids,
    attention_mask=attention_mask,
    max_length=100,  # Adjust the max length as needed
    num_return_sequences=1,
    no_repeat_ngram_size=2,  # To avoid repetitive phrases
    top_k=50,  # Consider the top k tokens
    top_p=0.95,  # Consider the cumulative probability up to p
    temperature=0.7,  # Control the creativity of the output
    do_sample=True,  # Enable sampling to generate more creative results
    pad_token_id=tokenizer.pad_token_id  # Set the pad token ID
)

# Decode the generated text to string format
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(generated_text[:60])
```
